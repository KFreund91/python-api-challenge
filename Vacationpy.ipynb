```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from config import g_key
```

```python
#Read in data and create DataFrame
weather_data = "cities.csv"
weather_data = pd.read_csv(weather_data)
weather_data.head()
```

```python
# Configure gmaps with API key
gmaps.configure(api_key=g_key)
```

```python
# Set Latitude and Longitude as locations and humidity as weight
humidity = weather_data["Humidity"].astype(float)
maxhumidity = humidity.max()
locations = weather_data[["Latitude", "Longitude"]]
```

```python
# Create first map
fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations, weights=humidity,dissipating=False, max_intensity=maxhumidity,point_radius=1)
fig.add_layer(heat_layer)
fig
```

```python
# Narrow down dataframe to reflect ideal conditions
ideal_temp = weather_data[(weather_data['Temperature'] > 70) & (weather_data['Temperature'] < 80)]
ideal_clouds = ideal_temp[(ideal_temp['Cloudiness'] == 0)]
ideal_df = ideal_clouds[(ideal_clouds['Wind Speed'] < 10)]
ideal_df
```
